```python
import pandas as pd 
import numpy as np 

# getting namUs Database
missing_db_1 = pd.read_csv("./created databases/ready-search-missing-person.csv")

missing_db_1.head()
```

```python
# last contacted input 
dlc = input("When was the last time you had contact/last seen loved one? MM/DD/YYYY : ")
# which state last seen input 
state = input("What state was your loved one last seen/known to be in?: ")
# age of when went missing 
age = input("What was the age of your loved one when first missing?: ")

def findSimilarCases(dlc, state, age): 

    # year of missing 
    dlc = dlc.split("/")[2]

    # gets cases in db that have same year as input year 
    similar_dlc_year = missing_db_1[missing_db_1["DLC"].str.split("/").str[2].fillna("") == dlc]

    # gets cases in db that have same year and state input 
    similar_dlc_state = similar_dlc_year[similar_dlc_year["State"] == state]

    # gets cases in db that have same year, state, and age input 
    similar_dlc_state.loc[:, 'Missing Age'] = pd.to_numeric(similar_dlc_state.loc[:, 'Missing Age'], errors='coerce')
    similar_dlc_state_age = similar_dlc_state[similar_dlc_state['Missing Age'] == pd.to_numeric(age)]

    # print cases that have same input values
    print(similar_dlc_state_age)

findSimilarCases(dlc, state, age)
```
